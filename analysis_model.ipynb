```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid", palette="Set2")
```

```python
df = pd.read_csv('data/demand_forecasting.csv')
```

```python
df
```

```python
df['Epidemic'].value_counts()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.describe().T
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Weekday'] = df['Date'].dt.day_name()
```

```python
df.head()
```

```python
df.columns
```

```python
df['Discounted Price'] = df['Price'] * (1 - df['Discount'] / 100)
```

```python
df.head()
```

```python
df['Sell Trough Rate'] = df['Units Sold'] / df['Inventory Level']
```

```python
df.head()
```

```python
df.groupby('Category')['Demand'].agg(['mean', 'sum', 'std']).sort_values(by='sum', ascending=False).reset_index()
```

```python
df.groupby(['Region', 'Seasonality'])['Demand'].mean().unstack()
```

```python
pd.pivot_table(df, values='Demand', index='Month', columns='Category', aggfunc="mean")
```

```python
sns.histplot(df['Demand'], bins=20, kde=True)
plt.title('Distribution of Demand')
plt.show()
```

```python
sns.scatterplot(df, x='Inventory Level', y='Units Sold')
plt.title('Unit Sold Vs Inventory Level')
plt.show()
```

```python
sns.boxplot(data=df, x='Category', y='Demand')
plt.title('Demand By Category')
plt.show()
```

```python
sns.boxplot(data=df, x='Weather Condition', y='Demand')
plt.title('Demand by Weather Condition')
plt.show()
```

```python
monthly_demand = df.groupby('Month')['Demand'].mean()
```

```python
monthly_demand
```

```python
monthly_demand.plot(kind='bar')
plt.title('Average Monthly Demand')
plt.show()
```

```python
daily_demarnd = df.groupby('Date')['Demand'].sum()
```

```python
daily_demarnd
```

```python
daily_demarnd.plot()
plt.title('Daily Demand Over Time')
plt.xlabel('Date')
plt.ylabel('Demand')
plt.show()
```

```python
df.columns
```

```python
sns.barplot(data=df, x='Promotion', y='Demand')
plt.title('Demand By Promotion')
plt.show()
```

```python
sns.scatterplot(data=df, x = 'Discounted Price', y = 'Demand')
plt.title('Discounted Price Vs Demand')
plt.show()
```

```python
df.groupby('Seasonality')['Demand'].mean().plot(kind='bar', title='Demand by seasonality')
plt.show()
```
